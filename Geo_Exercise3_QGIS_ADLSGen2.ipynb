# Azure Data Lake Storage Gen2 上のファイルを QGIS でロードする

Azure Data Lake Storage Gen2 上に保管されているラスターファイルやベクターファイルは一旦ローカルにダウンロードすることなく、直接 QGIS でロードすることができます。

クラウド上に保管されているファイルは、[GDAL 仮想ファイル システム](https://gdal.org/user/virtual_file_systems.html)の仕組みを利用してロードされます。Azure の Storage サービスの場合、以下のプレフィックスが利用可能です。
- `/vsiaz/` (Microsoft Azure Blob ファイル) **（QGIS の UI デフォルト）**<br>
ファイル全体を事前にダウンロードすることなく、Microsoft Azure Blob コンテナーで使用可能な (主に非公開の) ファイルをオンザフライでランダムに読み取ることができるファイル システム ハンドラーです。
- `/vsiaz_streaming/` (Microsoft Azure Blob ファイル: ストリーミング)
- `/vsiadls/`  (Microsoft Azure Data Lake Storage Gen2)<br>
ファイル全体を事前にダウンロードすることなく、Microsoft Azure Data Lake Storage ファイル システムで利用可能な (主に非公開の) ファイルをオンザフライでランダムに読み取ることができるファイル システム ハンドラーです。

## QGIS のインストール
[QGIS 公式サイト](https://qgis.org/ja/site/)より QGIS をダウンロードしてインストールを行います。

## 接続文字列の設定

1. QGIS を起動して、メニューから[設定]->[オプション]を選択します。

    ![QGIS Settings](https://github.com/nohanaga/databricks-geotools/raw/main/images/image001.png)

1. 左ペインの[システム]を選択し、[環境]に以下のカスタム変数を追加します。値(Value) には、Azure Data Lake Storage Gen2 の接続文字列を貼り付けます。

    `AZURE_STORAGE_CONNECTION_STRING`

1. Azure Data Lake Storage Gen2 の接続文字列は[Azure Portal](https://portal.azure.com/)から取得します。

    ![ADLS2](https://github.com/nohanaga/databricks-geotools/raw/main/images/image002.png)

1. Azure Data Lake Storage Gen2 リソースを開き、左メニューから[アクセスキー]を選択し、[接続文字列]を表示してからコピーボタンをクリックします。
1. クリップボードにコピーされた接続文字列を環境変数の値(Value)に貼り付けて[OK]をクリックします。
1. 設定を反映させるため、**QGIS を再起動**します。

## ファイルを読み込む

1. [Azure Storage Explorer](https://azure.microsoft.com/products/storage/storage-explorer/) もしくは、ストレージ ブラウザーを使用して読み込む GeoTIFF ファイルの場所を確認します。以下の例では、Blob コンテナー名 `data`、ファイル名 `63c3e7bc9a340e000522abae.tif` となります。

    ![Azure Storage Explorer](https://github.com/nohanaga/databricks-geotools/raw/main/images/image003.png)

1. QGIS を起動し、データソースマネージャを開き、以下のようにラスターファイルをレイヤーに追加します。

    ![Data Source Manager](https://github.com/nohanaga/databricks-geotools/raw/main/images/image004.png)

1. レイヤーにラスターファイルが追加され、ローカルと同じように扱うことができるようになります。
